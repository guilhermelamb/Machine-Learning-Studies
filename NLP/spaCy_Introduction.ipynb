```python
import spacy
```

```python
text = 'Em Linguística, a noção de texto é ampla e ainda aberta a uma definição mais precisa. Grosso modo, pode ser entendido como manifestação linguística das ideias de um autor, que serão interpretadas pelo leitor de acordo com seus conhecimentos linguísticos e culturais. Seu tamanho é variável. Todo texto tem que ter alguns aspectos formais, ou seja, tem que ter estrutura, elementos que estabelecem relação entre si. Dentro dos aspectos formais temos a coesão e a coerência, que dão sentido e forma ao texto. "A coesão textual é a relação, a ligação, a conexão entre as palavras, expressões ou frases do texto”'
```

```python
nlp = spacy.load('pt_core_news_lg')
doc = nlp(text)
```

```python
#tokens

tokens = [token for token in doc]
print(tokens)
```

```python
tokens = [token.orth_ for token in doc]
print(tokens)
```

```python
tokens = [token.orth_ for token in doc if token.is_alpha]
print(tokens)
```

```python
tokens = [token.orth_ for token in doc if token.is_digit]
print(tokens)
```

```python
tokens = [token.orth_ for token in doc if token.is_punct]
print(tokens)
```

```python
#Lemmatization

lemma = [token.lemma_ for token in doc if token.pos_ == 'VERB']
print(lemma)
```

```python
#tags

tags = [(token.orth_, token.pos_) for token in doc]
print(tags)
```

```python
tags = [(token.orth_, token.morph) for token in doc]
print(tags)
```

```python
#named entities

entities = list(doc.ents)
print(entities)
```

```python
entities_detail = [(entity, entity.label_) for entity in doc.ents]
print(entities_detail)
```

```python
from pathlib import Path

html = spacy.displacy.render(doc, style='ent')
output_path = Path('named_entities.html')
output_path.open('w', encoding='utf-8').write(html)
```

```python
#syntax analisys

syntax = [(token.orth_, token.dep_) for token in doc]
print(syntax)
```

```python
html_2 = spacy.displacy.render(doc, style='dep')
output_path = Path('syntax_analisys.html')
output_path.open('w', encoding='utf-8').write(html_2)
```
